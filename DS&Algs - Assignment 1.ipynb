## Task 2 - Factorial Algorithm Implementations
For different values of n, implement your algorithm using two approaches:  iterative and recursive (tail and non-tail recursions).  For each case, plot running time as a function of n and compare the achieved result.

```python
def get_time_list(func):
    '''A function to takes a given function and 
    returns a list containing the times for different values of n'''
    time = []
    input_size = []
    for i in range(1, 1500, 100): # from 0 to 1500 in steps of 100 (15 values)
        foo = %timeit -n1000 -r6 -o func(i) # r runs, n loops of each
        time.append(foo)
        
        input_size.append(i)

    time_values = []
    for i in range(len(time)):
        time_values.append(time[i].timings)
        
    return time_values, input_size

def get_mean_times(time_list):
    '''A function to get the MEAN times 
    from the set of times generated in the get_time_list func.'''
    mean_times = []
    for i in time_list:
        mean_times.append(sum(i)/len(i))
    return mean_times

def get_min_times(time_list):
    '''A function to get the MINIMUM times 
    from the set of times generated in the get_time_list func.'''
    min_times = []
    for i in time_list:
        min_times.append(min(i))
    return min_times



```

<b><i>2.1 Non-Tail Recursive</i></b>

```python
def fact_recursive_non_tail(n):
    '''Implementation of the Factorial algorithm 
    using the Recursive Non-Tail approach'''
    if (n == 0):
        return 1 # base case
    return n * fact_recursive_non_tail(n-1)
```

```python
time_list1 = get_time_list(fact_recursive_non_tail)
input_size1 = list(time_list1[1]) # list of inputs is the second element returned by the get_time_list() function
time_list1 = list(time_list1[0]) # list of running times is the first element returned by the get_time_list() function
```

```python
mean_times1 = get_mean_times(time_list1)
min_times1 = get_min_times(time_list1)

print(mean_times1)
print(input_size1)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.set_xlabel('Input')
ax.set_ylabel('Time [s]')
ax.set_title('Non-Tail Recursive')
plt.plot(input_size1, mean_times1)
plt.show()
```

<b><i>2.2 Tail Recursive</i></b>

```python
def fact_recursive_tail(n, a = 1):
    '''Implementation of the Factorial algorithm 
    using the Tail Recursive approach'''
    if (n == 0):
        return a # base case
    return fact_recursive_tail(n-1, a*n);
```

```python
time_list2 = get_time_list(fact_recursive_tail)
input_size2 = list(time_list2[1]) # list of inputs is the second element returned by the get_time_list() function
time_list2 = list(time_list2[0]) # list of running times is the first element returned by the get_time_list() function
```

```python
mean_times2 = get_mean_times(time_list2)
min_times2 = get_min_times(time_list2)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.set_xlabel('Input')
ax.set_ylabel('Time [s]')
ax.set_title('Tail Recursive')
plt.plot(input_size2, mean_times2)
plt.show()
```

<b><i>2.3 Iterative</i></b>

```python
def fact_iter(n):
    '''Implementation of the Factorial algorithm 
    using the Iterative approach'''
    if (n == 0):
        return 1
    else:
        factorial = 1
        for i in range(1, n+1):
            factorial *= i
        return factorial
```

```python
time_list3 = get_time_list(fact_iter)
input_size3 = list(time_list3[1]) # list of inputs is the second element returned by the get_time_list() function
time_list3 = list(time_list3[0]) # list of running times is the first element returned by the get_time_list() function
```

```python
mean_times3 = get_mean_times(time_list3)
min_times3 = get_min_times(time_list3)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.set_xlabel('Input')
ax.set_ylabel('Time [s]')
ax.set_title('Iterative')
plt.plot(input_size3,mean_times3)
plt.show()
```

<b><i>3.1 Linear search</i></b>

```python
def linearSearch(sequence, target):
    for i in sequence:
        if i == target:
            return sequence.index(i)
    return -1
```

<b><i>3.3 Binary search - Iterative</i></b>

```python
def binarySearchIterative(sequence, target, left = 0, right = len(sequence)-1): 
    
    while left <= right: 
  
        mid = (right + left) // 2;
          
        # Check if target is present at mid 
        if sequence[mid] == target: 
            return mid 
  
        # If target is greater, ignore left half 
        if sequence[mid] < target: 
            left = mid + 1
  
        # If target is smaller, ignore right half 
        elif sequence[mid] > target: 
            right = mid - 1
      
    # If we reach here, then the element 
    # was not present 
    return -1
```

## Task 4 - Best and Worst Running Times for Search Algorithms

<b>Functions:</b>
<br>Below is a collection of functions that will be used to compare the running times of the algorithms discussed in Task 3. 

```python
def get_time_list_search(func, sequence, target):
    '''
    A function that calculates the running time of a given Search Algorithm function.
    Returns a list of lists, each inner list containing r average running times (one for each run).'''
    time = [] # is the time array growing dynamically causing spikes and troughs
    foo = %timeit -n100 -r10 -o func(sequence, target) # r runs, n loops of each
    time.append(foo)

    time_values = []
    for i in range(len(time)):
        time_values.append(time[i].timings)

    return time_values

    
# def get_mean_times_search(time_list):
#     '''A function that gets the mean times from a list '''
#     mean_times = []
#     for i in time_list:
#         mean_times.append(sum(i)/len(i))
#     return mean_times


def get_min_times_search(time_list):
    '''A function that gets the minimum times from a list'''
    min_times = []
    for i in time_list:
        min_times.append(min(i))
    return min_times

# A function to get a list of running times for a given function
# 
def get_running_times(function, sequence, target, min_or_mean = 'min', case = None, upperbound = 600):
    '''
    A function to get a list of running times for a given function. 
    Also returns a list of equal size, containing the corresponding input size
    '''
#     sequence.sort() # necessary for binary search
    times = []
    input_size = []
    timelist = []
    
    
    while len(sequence) < upperbound: # upperbound = argument passed to function for number of loops
        timelist = get_time_list_search(function, sequence, target)
        input_size.append(len(sequence))
        
        if min_or_mean == 'mean':
            t = get_mean_times_search(timelist)
        elif min_or_mean == 'min':
            t = get_min_times_search(timelist)

        times.append(t[0])
        
        if function == binarySearchIterative:
#             for i in range(5):
                #sequence.append(sum(sequence))
            sequence.append(sequence[-1]+1)
            
            if case == 'Best':
                target = get_middle_element(sequence) # to keep the target always as the mid value of the sequence
            elif case == 'Worst':
                target = -5 # number not in the list (all numbers in sequence are +ve)
                    
        elif function == linearSearch:
            sequence.append(sum(sequence))
            if case == 'Worst':
                target = sequence[-1] # to keep the target always as the last element of the sequence
            
    return times, input_size



def get_middle_element(sequence):
    '''For Binary Search: 
    Finding the mid element in the sequence. 
    (In the same way the binarySearchIterative function would)'''
    left = 0
    right = len(sequence)-1
    mid = (right + left) // 2
    return sequence[mid]
```

## Task 5 - Curve Fitting

```python
def plot_fitted_line(x, y, labels = None, title = None):
    '''Function to plot the fitted line for one or multiple scatter plots'''
    fig, ax = plt.subplots()
    ax.set_xlabel('Input size')
    ax.set_ylabel('Time [s]')
    ax.set_title(title)
    for i in range(len(x)): # for multiple plots
        res = stats.linregress(x[i], y[i])
        slope = res.slope
        intercept = res.intercept
        if labels == None:
            plt.plot(x[i], y[i], 'o')
            plt.plot(np.asarray(x[i]), res.intercept + res.slope*np.asarray(x[i]), label='Fitted line')
            print('Equation is: T(n) = {:.3g} * n + {:.3g}'.format(slope, intercept))
        else:
            plt.plot(x[i], y[i], 'o', label=labels[i])
            plt.plot(np.asarray(x[i]), res.intercept + res.slope*np.asarray(x[i]), label='Fitted line for {}'.format(labels[i]))
            print('Equation for "{}" is: T(n) = {:.3g} * n + {:.3g}'.format(labels[i], slope, intercept))
        
    plt.legend()
    plt.show()


    return slope, intercept
```
